Algoritmo para calcular la factura de energia eléctrica
1. iniciar
2. definir las variables ("consumo(Whk)", "costo Whk(pesos colombianos), "estrato(1-6)", altitud(m sobre nivel del mar)")
3. calcular el consumo básico según la altitud, si la altitud >1000 el consumo basico = 130 kWh , si la altitus es <=1000 el consumo basico = 173 kWh
4. calcular el subsidio o contribucion segun el estrato (estrato 1, subsidio 60% del CU para el Consumo Básico), (estrato 2, subsidio = 50% del CU para el Consumo Básico) , (estrato 3, subsidio = 15% del CU para el Consumo Básico), (estrato 4, sin subsidio ni contribución), (Si estrato 5 o 6, contribución = 20% del CU en todo el consumo), ( usuario es comercial, contribución = 20% del CU en todo el consumo)
5. calcular el valor de la factura (consumo <= Consumo Básico, valor_factura = (consumo x costo_kwh) - subsidio) , (Si consumo > Consumo Básico, valor_factura = (Consumo Básico x costo_kwh) - subsidio + ((consumo - Consumo Básico) x costo_kwh) + contribución)
6. finalizar

```python
def calcular_factura(consumo, costo_kwh, estrato, altitud):
    # Calcular Consumo Básico
    if altitud > 1000:
        consumo_basico = 130
    else:
        consumo_basico = 173

    # Calcular subsidio o contribución
    if estrato == 1:
        subsidio = 0.6 * costo_kwh * consumo_basico
    elif estrato == 2:
        subsidio = 0.5 * costo_kwh * consumo_basico
    elif estrato == 3:
        subsidio = 0.15 * costo_kwh * consumo_basico
    elif estrato == 4:
        subsidio = 0
    elif estrato >= 5:
        contribucion = 0.2 * costo_kwh * consumo
    else:
        contribucion = 0

    # Calcular valor de la factura
    if consumo <= consumo_basico:
        valor_factura = (consumo * costo_kwh) - subsidio
    else:
        valor_factura = (consumo_basico * costo_kwh) - subsidio + ((consumo - consumo_basico) * costo_kwh) + contribucion

    return valor_factura
```
